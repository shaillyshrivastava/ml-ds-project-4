## Adding Average Growth Rate as new Column

```python
state_mean={}
for column in gdp_current_price:
    if column in states:
        mean=gdp_current_price[gdp_current_price["Items Description"]=="(% Growth over previous year)"][column].mean()
        state_mean[column]=mean
state_mean["Items Description"]="Average growth rate"
gdp_current_price=gdp_current_price.append(state_mean,ignore_index=True)
```

```python
import os
```

```python
listOfFiles=os.listdir(path)
```

```python
len(listOfFiles)
```

```python
len(states)
```

```python
for filename in listOfFiles:
    for statename in states:
        if statename in filename:
            filepath=path+filename
            data=pd.read_csv(filepath,encoding="ISO-8859-1")
            data=data[['Item','2014-15']]
            data=data.T
            data.columns=data.iloc[0]
            data=data[1:]
            data["state"]=statename
            gsva=pd.concat([data,gsva])
```

```python
# Scatter Plot for GDP per capita with dropout rates in education

f = plt.figure()    
f, axes = plt.subplots(nrows = 1, ncols = 2, sharex=True, sharey = False)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=None)

sc = axes[0].scatter(gsva_primary['Primary - 2014-2015'],gsva_primary['Per Capita GSDP (Rs.)'], s=100, c='DarkRed',marker="o")
axes[0].set_ylabel('Per Capita GSDP (Rs.)')
axes[0].set_xlabel('Primary Education')

sc = axes[1].scatter(gsva_secondary['Secondary - 2014-2015'],gsva_secondary['Per Capita GSDP (Rs.)'], s=100, c='DarkGreen',marker="s")
axes[1].set_ylabel('Per Capita GSDP (Rs.)')
axes[1].set_xlabel('Secondary Education')
```

1. Drop out rates are higher in Secondary Education.
2. Drop out rates for Primary are less than or equal to 15.
3. Drop out rates for Secondary is generally above than 5.
4. Statistics for Primary and Upper Primary are pretty similar.
5. The outliers are due to Goa and Sikkim because of its Per Capita GDP
